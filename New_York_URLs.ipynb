```python
from selenium import webdriver
```

```python
driver = webdriver.Chrome(executable_path='C:\\chromedriver')
```

```python
length = 161
# you can incease this in order to get more data.
```

```python
def getData(url):    
    driver.get(url)    
    topic_list = driver.find_elements_by_class_name("entry-heading")    
    sub_link_list = []
    for topic in topic_list:
        t_list = topic.find_elements_by_xpath(".//*")
        for element in t_list:
            if element.tag_name=="a":
                sub_link_list.append(element.get_attribute('href'))
    return sub_link_list
```

```python
urls = []
for i in range(length):
    url = getData("https://nypost.com/search/trump/opinion/page/"+str(i+1)+"/")
    urls.append(url)
```

```python
final_urls = []
for i in range(length):
    for j in range(10):
        final_urls.append(urls[i][j])
```

```python
urls_str = ""
for url in final_urls:
    urls_str += str(url) +"\n"
urls_str.strip("\n")
```

```python
with open("opinionfinal_New_York_url.txt", "w", encoding="utf-8") as file:
    file.write(urls_str)
```
